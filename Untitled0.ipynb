```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt


connection = sqlite3.connect("/content/winequality-white.sqlite")
cursor = connection.cursor()
print(" ·Éë·Éê·Éñ·Éê·É°·Éó·Éê·Éú ·Éô·Éê·Éï·É®·Éò·É†·Éò ·Éì·Éê·Éõ·Éß·Éê·É†·Éì·Éê")

cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()
if not tables:
    print(" ·Éë·Éê·Éñ·Éê ·É™·Éê·É†·Éò·Éî·Éö·Éò·Éê, ·Éï·Éô·Éò·Éó·ÉÆ·É£·Éö·Éù·Éë·Éó CSV ·É§·Éê·Éò·Éö·É° ·Éì·Éê ·Éï·É•·Éõ·Éú·Éò·Éó ·É™·ÉÆ·É†·Éò·Éö·É°...")
    df = pd.read_csv("/content/winequality-white new.csv")
    df.to_sql("wine", connection, if_exists="replace", index=False)
    print(" CSV ·É¨·Éê·É†·Éõ·Éê·É¢·Éî·Éë·Éò·Éó ·Éí·Éê·Éì·Éê·Éò·É¢·Éê·Éú·Éê ·Éë·Éê·Éñ·Éê·É®·Éò ·É†·Éù·Éí·Éù·É†·É™ 'wine' ·É™·ÉÆ·É†·Éò·Éö·Éò")
else:
    print(" ·Éë·Éê·Éñ·Éê·É®·Éò ·Éê·É†·É°·Éî·Éë·É£·Éö·Éò ·É™·ÉÆ·É†·Éò·Éö·Éî·Éë·Éò·Éê:", tables)

print("\n--- ·É¶·Éï·Éò·Éú·Éù·Éî·Éë·Éò ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò·Éó 7 ---")
query = "SELECT * FROM wine WHERE [quality] = 7"
cursor.execute(query)
results = cursor.fetchall()
for row in results[:10]:  # 10 ·Éõ·Éê·Éí·Éê·Éö·Éò·Éó·Éò
    print(row)

print("\n--- ·Éê·ÉÆ·Éê·Éö·Éò ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò·É° ·Éì·Éê·Éõ·Éê·É¢·Éî·Éë·Éê ---")
fixed = float(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó fixed acidity: "))
volatile = float(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó volatile acidity: "))
citric = float(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó citric acid: "))
quality = int(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò (0-10): "))

cursor.execute("""
    INSERT INTO wine ([fixed acidity], [volatile acidity], [citric acid], [quality])
    VALUES (?, ?, ?, ?)
""", (fixed, volatile, citric, quality))
connection.commit()
print(" ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò ·Éì·Éê·Éî·Éõ·Éê·É¢·Éê ·Éë·Éê·Éñ·Éê·É®·Éò!")

print("\n--- ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò·É° ·Éí·Éê·Éú·Éê·ÉÆ·Éö·Éî·Éë·Éê ---")
old_quality = int(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·É´·Éï·Éî·Éö·Éò ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò, ·É†·Éù·Éõ·Éö·Éò·É° ·É®·Éî·É™·Éï·Éö·Éê·É™ ·Éí·É°·É£·É†·Éó: "))
new_quality = int(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·Éê·ÉÆ·Éê·Éö·Éò ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò: "))

cursor.execute("UPDATE wine SET [quality] = ? WHERE [quality] = ?", (new_quality, old_quality))
connection.commit()
print("‚úèÔ∏è ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò ·Éí·Éê·Éú·Éê·ÉÆ·Éö·Éì·Éê!")


print("\n--- ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò·É° ·É¨·Éê·É®·Éö·Éê ---")
quality_to_delete = int(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò, ·É†·Éù·Éõ·Éö·Éò·É° ·É¨·Éê·É®·Éö·Éê·É™ ·Éí·É°·É£·É†·Éó: "))
cursor.execute("DELETE FROM wine WHERE [quality] = ?", (quality_to_delete,))
connection.commit()
print("üóëÔ∏è ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò ·É¨·Éê·Éò·É®·Éê·Éö·Éê!")


print("\n--- ·É¶·Éï·Éò·Éú·Éò·É° ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò·É° ·Éí·Éê·Éú·Éê·É¨·Éò·Éö·Éî·Éë·Éò·É° ·Éì·Éò·Éê·Éí·É†·Éê·Éõ·Éê ---")
df = pd.read_sql_query("SELECT [quality] FROM wine", connection)

plt.figure(figsize=(8,5))
df['quality'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title("üç∑ ·É¶·Éï·Éò·Éú·Éò·É° ·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò·É° ·Éí·Éê·Éú·Éê·É¨·Éò·Éö·Éî·Éë·Éê")
plt.xlabel("·ÉÆ·Éê·É†·Éò·É°·ÉÆ·Éò")
plt.ylabel("·É†·Éê·Éù·Éì·Éî·Éú·Éù·Éë·Éê")
plt.grid(True)
plt.show()

connection.close()
print(" ·Éë·Éê·Éñ·Éê·É°·Éó·Éê·Éú ·Éô·Éê·Éï·É®·Éò·É†·Éò ·Éì·Éê·Éò·ÉÆ·É£·É†·Éê")

```
